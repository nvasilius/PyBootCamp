```python
x=1
while x<=100:
    if x%3==0 and x%5==0:
        print('FizzBuzz')
    elif x%3==0:
        print('fizz')
    elif x%5==0:
        print('buzz')
    else:
        print(x)
    x+=1


testString='racecar'
if testString == testString[::-1]:
    print('Yay, Palindrome!')
else:
    print('Nope, not a Palindrome.')
```
